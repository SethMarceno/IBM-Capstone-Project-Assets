Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FRmOM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
query = '''
select ConvertedComp
from master
where ConvertedComp is not Null
'''

df = pd.read_sql_query(query, conn)

count, bins = np.histogram(df, 10)

df.plot(kind='hist', figsize=(14,7), xticks=bins)

plt.title('Histogram of ConvertedComp')
plt.ylabel('Frequency')
plt.xlabel('Converted Compensation in USD')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
query = '''
SELECT Age 
from master
where Age is not Null
'''

df = pd.read_sql_query(query, conn)

df.plot(kind='box', figsize=(10, 15))

plt.title('Boxplot of Age')
#plt.ylabel('Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query = '''
select Age, WorkWeekHrs
from master
where Age is not Null and WorkWeekHrs is not Null
'''

df=pd.read_sql_query(query, conn)

df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(14, 7))

plt.title('Scatter Plot of Age vs Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
query='''
select WorkWeekHrs, CodeRevHrs, Age
from master
where WorkWeekHrs is not Null and CodeRevHrs is not Null and Age is not Null
'''

df = pd.read_sql_query(query, conn)
norm_Age = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())

df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', figsize=(20, 6), 
        alpha=0.5, color='blue', s=norm_Age * 350)

plt.title('Bubble plot of Work Week Hours vs Code Revision Hours with Age as weights')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Revision Hours')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query='''
select *
from DatabaseDesireNextYear
'''

df = pd.read_sql_query(query, conn)
df2 = df['DatabaseDesireNextYear'].value_counts()
df_top5 = df2.head()

df_top5.plot(kind='pie', figsize=(6, 6),
                        autopct='%1.1f%%', 
                        startangle=90,    
                        shadow=True,           
                        )

plt.title('Top 5 databases desired next year')
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
query='''
select WorkWeekHrs, CodeRevHrs, Age
from master 
where WorkWeekHrs is not Null and CodeRevHrs is not Null and Age between 30 and 35
'''

df = pd.read_sql_query(query, conn)
df.sort_values('Age', ascending=True, inplace=True)

med = df.groupby('Age').median()

med.plot(kind='area', alpha=0.3, stacked=True, figsize=(14,7))

plt.title('Stacked Chart of median Work Week Hours and Code Revision Hours by age 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
query='''
select ConvertedComp, Age
from master
where ConvertedComp is not Null and Age between 45 and 60
'''
df = pd.read_sql_query(query, conn)
df.sort_values('Age', ascending=True, inplace=True)
med = df.groupby('Age').median()

med.plot(kind='line', figsize=(14, 7))

plt.title('Median Converted Compensation in USD for ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Converted Compensation in USD')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
query='''
select MainBranch
from master
where MainBranch is not Null
'''

df = pd.read_sql_query(query, conn)
counts = df['MainBranch'].value_counts()

counts.plot(kind='barh', figsize=(14, 7))

plt.title('Bar chart of Main Branch')
plt.xlabel('Count')

plt.show()
```
