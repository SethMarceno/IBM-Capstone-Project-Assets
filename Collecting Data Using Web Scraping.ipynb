## Extract information from the given web site

You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
from bs4 import BeautifulSoup 
import requests
```

Download the webpage at the url


```python
#your code goes here
data = requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data, 'html5lib')
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
import pandas as pd

table = soup.find('table')

Language = []
Avg_Ann_Sal = []
for row in table.find_all('tr'): #rows
    
    cols = row.find_all('td') 
    Language.append(cols[1].getText()) 
    Avg_Ann_Sal.append(cols[3].getText())
Language.remove('Language')
Avg_Ann_Sal.remove('Average Annual Salary')
AAS = []

for entry in Avg_Ann_Sal:
    new = entry.replace(',', '')
    AAS.append(int(new[1:]))

    
print(AAS)    
    
dat = {'Language': Language, 'Average Annual Salary': AAS}

df = pd.DataFrame(dat)
df.sort_values(by='Average Annual Salary', axis=0, ascending=False, inplace=True)
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='bar', x='Language', figsize=(14, 7))

plt.title('Average salary for the Most Popular Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Average Salary (USD)')

plt.show()
```

Save the scrapped data into a file named _popular-languages.csv_


```python
# your code goes here
df.to_csv(r'C:\users\seth marceno\desktop\popular-languages.csv', index = False)
```
